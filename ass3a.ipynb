```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('adult.csv')
df
```

```python
df.shape
```

```python
# empty dataframe to store the calculated statistics
results = pd.DataFrame()
```

```python
newdf = df.select_dtypes(include = 'number')
```

```python
for i in newdf.columns:
    mean = newdf[i].mean()
    median = newdf[i].median()
    mode = newdf[i].mode()
    minValue = newdf[i].min()
    maxValue = newdf[i].max()
    std = newdf[i].std()
    variance = newdf[i].var()
    percentile_25 = newdf[i].quantile(0.25)
    percentile_50 = newdf[i].quantile(0.50)
    percentile_75 = newdf[i].quantile(0.75)
    results[i] = [mean, median, mode, minValue, maxValue, std, variance, percentile_25, percentile_50, percentile_75]
```

```python
results.index = ['Mean', 'Median', 'Mode', 'Min', 'Max', 'Standard Deviation', 'Variance', '25%', '50%', '75%']
```

```python
print(results)
```
